## 1. Introduction & Problem Statement
Understand how to read, summarize, and visualize a dataset using the classic Iris flower dataset.

**Dataset:** Iris Dataset (150 samples, 3 species, 4 numeric features)  
**Goal:** Practice data loading, summarization, and visualization fundamentals.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style='whitegrid')
print('✅ Libraries imported')
```

## 2. Dataset Understanding

```python
df = pd.read_csv('iris.csv')
print('Shape:', df.shape)
print('Columns:', df.columns.tolist())
df.head()
```

```python
df.info()
print()
df.describe()
```

## 3. Data Cleaning & Preparation
Iris dataset is already clean — verifying below.

```python
print('Missing values:\n', df.isnull().sum())
print('\nDuplicate rows:', df.duplicated().sum())
```

## 4. Exploratory Data Analysis (EDA)

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species', s=70)
plt.title('Scatter Plot: Sepal Length vs Petal Length')
plt.savefig('scatter_plot.png', dpi=150)
plt.show()
```

```python
df.hist(figsize=(10,7), bins=20, edgecolor='black')
plt.suptitle('Histograms of Feature Distributions')
plt.tight_layout()
plt.savefig('histograms.png', dpi=150)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df.drop(columns='species'))
plt.title('Box Plot — Outlier Detection')
plt.savefig('boxplot.png', dpi=150)
plt.show()
```

## 5. Conclusion
- Dataset is clean: no missing values, no duplicates.
- Petal length and width strongly correlate and separate species well.
- Setosa is clearly distinct; Versicolor/Virginica overlap slightly.
- Very few outliers present — dataset is balanced (50 samples/species).
